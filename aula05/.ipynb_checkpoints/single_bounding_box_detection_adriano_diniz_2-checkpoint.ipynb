## Dataset

O dataset escolhido foi o [Street View Hous Numbers Dataset](http://ufldl.stanford.edu/housenumbers/), contendo os dados já pré-separados em pastas de treinamento e teste.

```python
# Transform classes adapted from PyTorch Data Loading tutorial:
# https://pytorch.org/tutorials/beginner/data_loading_tutorial.html

class Rescale(object):
    """Rescale the image in a sample to a given size.

    Args:
        output_size (tuple or int): Desired output size. If tuple, output is
            matched to output_size. If int, smaller of image edges is matched
            to output_size keeping aspect ratio the same.
    """

    def __init__(self, output_size):
        assert isinstance(output_size, tuple)
        self.output_size = output_size

    def __call__(self, sample):
        image, bbox = sample['image'], sample['bbox']

        h, w = image.shape[:2]
        if isinstance(self.output_size, int):
            if h > w:
                new_h, new_w = self.output_size * h / w, self.output_size
            else:
                new_h, new_w = self.output_size, self.output_size * w / h
        else:
            new_h, new_w = self.output_size

        new_h, new_w = int(new_h), int(new_w)

        img = transform.resize(image, (new_h, new_w))

        ratio_w = new_w / w
        ratio_h = new_h / h
        bbox = [bbox[0] * ratio_w, bbox[1] * ratio_h, bbox[2] * ratio_w, bbox[3] * ratio_h]

        return {'image': img, 'bbox': bbox, 'label': sample['label']}


class Normalize(object):
    """Normalize image and bounding box arrays."""

    def __call__(self, sample):
        h, w = sample['image'].shape[:2]
        bbox = np.array([sample['bbox'][0] / w,
                         sample['bbox'][1] / h,
                         sample['bbox'][2] / w,
                         sample['bbox'][3] / h])
        label = np.array([sample['label']])

        return {'image': sample['image'],
                'bbox': bbox,
                'label': label}

class ToTensor(object):
    """Convert ndarrays in sample to Tensors."""

    def __call__(self, sample):
        # swap color axis because
        # numpy image: H x W x C
        # torch image: C X H X W
        image = sample['image'].transpose((2, 0, 1))

        return {'image': torch.from_numpy(image).float(),
                'bbox': torch.from_numpy(sample['bbox']).float(),
                'label': torch.from_numpy(sample['label'])}
```

```python
class SVHNDataset(torch.utils.data.Dataset):
    def __init__(self, rootdir, train=True, transform=None, data_split=None, label_count=True, label_range=None):
        self.transform = transform

        # Get list of images
        path = os.path.join(rootdir, 'train' if train else 'test')
        file_list = [file_name for file_name in os.listdir(path) if file_name.endswith('.png')]

        # Update list of images to desired range
        if isinstance(data_split, int):
            if data_split > 0:
                file_list = file_list[:data_split]
            else:
                file_list = file_list[data_split:]
        elif isinstance(data_split, float):
            if data_split > 0.0:
                file_list = file_list[:int(np.ceil(len(file_list) * data_split))]
            else:
                file_list = file_list[int(np.ceil(len(file_list) * data_split)):]

        # Read bounding box and label data
        file_data = None
        for entry in os.listdir(path):
            if entry.endswith('.json'):
                file_data_obj = open(os.path.join(path, entry), 'r')
                file_data = json.load(file_data_obj)
                file_data_obj.close()
                break

        # Make list of image paths and merge bounding boxes
        self.file_path_list = list()
        self.file_data_list = list()
        for file_name in file_list:
            if (label_range is not None) and (len(file_data[file_name]) > label_range):
                continue

            self.file_path_list.append(os.path.join(path, file_name))

            entry = file_data[file_name][0]
            label_value = (str(entry['label']) if entry['label'] != 10 else '0')

            bbox_col_min = entry['left']
            bbox_row_min = entry['top']
            bbox_col_max = entry['left'] + entry['width']
            bbox_row_max = entry['top'] + entry['height']
            for idx in range(1, len(file_data[file_name])):
                entry = file_data[file_name][idx]
                label_value += (str(entry['label']) if entry['label'] != 10 else '0')

                bbox_col = entry['left'] + entry['width']
                bbox_row = entry['top'] + entry['height']

                if bbox_col_min > entry['left']:
                    bbox_col_min = entry['left']
                if bbox_row_min > entry['top']:
                    bbox_row_min = entry['top']
                if bbox_col_max < bbox_col:
                    bbox_col_max = bbox_col
                if bbox_row_max < bbox_row:
                    bbox_row_max = bbox_row

            self.file_data_list.append(dict(bbox=[bbox_col_min, bbox_row_min,
                                                  (bbox_col_max - bbox_col_min),
                                                  (bbox_row_max - bbox_row_min)],
                                            label=(len(label_value) if label_count else label_value)))

    def __len__(self):
        return len(self.file_path_list)

    def __getitem__(self, index):
        if type(index) is int:
            fp = self.file_path_list[index]
            fd = self.file_data_list[index]
        else:
            fp = self.file_path_list[index][0]
            fd = self.file_data_list[index][0]

        img = cv2.cvtColor(cv2.imread(fp, cv2.IMREAD_UNCHANGED+cv2.IMREAD_ANYDEPTH+cv2.IMREAD_ANYCOLOR).astype(np.float32)/255, cv2.COLOR_BGR2RGB)

        sample = {'image': img, 'bbox': fd['bbox'], 'label': fd['label']}

        if self.transform:
            sample = self.transform(sample)

        return sample
```

## Dataloaders para treinamento

```python
batch_size = 8

train_transform = torchvision.transforms.Compose([
    Rescale((60, 120)),
    Normalize(),
    ToTensor()
])
test_transform = torchvision.transforms.Compose([
    Rescale((60, 120)),
    Normalize(),
    ToTensor()
])

dataset_train = SVHNDataset(dataset_path, train=True, data_split=0.7, label_range=4, transform=train_transform)
dataset_valid = SVHNDataset(dataset_path, train=True, data_split=-0.3, label_range=4, transform=train_transform)
dataset_test = SVHNDataset(dataset_path, train=False, label_range=4, transform=test_transform)

dataloader_train = torch.utils.data.DataLoader(dataset_train, batch_size=batch_size, shuffle=True, num_workers=0)
dataloader_valid = torch.utils.data.DataLoader(dataset_valid, batch_size=batch_size, shuffle=False, num_workers=0)
dataloader_test = torch.utils.data.DataLoader(dataset_test, batch_size=batch_size, shuffle=False, num_workers=0)
```

## Definição da rede

```python
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=(3, 3))
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=(3, 3))
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=96, kernel_size=(3, 3))
        self.conv4 = nn.Conv2d(in_channels=96, out_channels=128, kernel_size=(3, 3))
        self.conv5 = nn.Conv2d(in_channels=128, out_channels=160, kernel_size=(3, 3))
        self.conv6 = nn.Conv2d(in_channels=160, out_channels=192, kernel_size=(3, 3))
        self.conv7 = nn.Conv2d(in_channels=192, out_channels=224, kernel_size=(3, 3))
        self.conv8 = nn.Conv2d(in_channels=224, out_channels=256, kernel_size=(3, 3))
        self.conv9 = nn.Conv2d(in_channels=256, out_channels=288, kernel_size=(3, 3))
        self.maxp = nn.MaxPool2d(kernel_size=(2, 2))

        self.dens1 = nn.Linear(in_features=288*2*17, out_features=64)
        self.drop1 = nn.Dropout(0.5)
        self.dens2 = nn.Linear(in_features=64, out_features=4)

    def forward(self, x):        # input:  amostras x  3  x 60 x 120
        x = self.conv1(x)        # output: amostras x 32  x 58 x 118
        x = F.relu(x)
        x = self.conv2(x)        # output: amostras x 64  x 56 x 116
        x = F.relu(x)
        x = self.maxp(x)         # output: amostras x 64  x 28 x 58

        x = self.conv3(x)        # output: amostras x 96  x 26 x 56
        x = F.relu(x)
        x = self.conv4(x)        # output: amostras x 128 x 24 x 54
        x = F.relu(x)
        x = self.maxp(x)         # output: amostras x 128 x 12 x 27

        x = self.conv5(x)        # output: amostras x 160 x 10 x 25
        x = F.relu(x)
        x = self.conv6(x)        # output: amostras x 192 x  8 x 23
        x = F.relu(x)
        x = self.conv7(x)        # output: amostras x 224 x  6 x 21
        x = F.relu(x)
        x = self.conv8(x)        # output: amostras x 256 x  4 x 19
        x = F.relu(x)
        x = self.conv9(x)        # output: amostras x 288 x  2 x 17
        x = F.relu(x)

        x = x.view(-1, 288*2*17) # output: amostras x (64 * 13 * 5)
        x = self.dens1(x)        # output: amostras x 64
        x = F.relu(x)
        x = self.drop1(x)
        x = self.dens2(x)        # output: amostras x 4

        return x.view(-1, 4)     # output: amostras x 1 x 4
```

## Instanciando o modelo que usará aumento sintético de dados

```python
model = MyModel()
model.to(device)
```

```python
dataiter = iter(dataloader_train)
samples = dataiter.next()
print(f'Dimensão da entrada: {samples["image"].shape}')
```

```python
samples_output = model.forward(samples['image'].to(device))
print(f'Dimensão da saída: {samples_output.shape}')
```

## Treinamento

```python
n_epochs = 200

crit = nn.MSELoss()
optm = torch.optim.Adam(params=model.parameters(), lr=1.e-4, betas=(0.9, 0.999))
schd = torch.optim.lr_scheduler.StepLR(optm, step_size=10, gamma=0.75)
```

```python
num_train_samples = len(dataloader_train.dataset)
num_train_batches = num_train_samples // batch_size
num_valid_samples = len(dataloader_valid.dataset)
num_valid_batches = num_valid_samples // batch_size

batch_print_size = 10

epoch_train_loss_history = list()
epoch_valid_loss_history = list()
print('Start training for {} epochs'.format(n_epochs))

try:
    for epoch in range(n_epochs):  # loop over the dataset multiple times
        train_t0 = time.time()
        epoch_train_loss = 0.0
        running_loss = 0.0

        for i, samples in enumerate(dataloader_train):
            # get the inputs
            inputs, labels = samples['image'], samples['bbox']
            inputs, labels = inputs.to(device), labels.to(device)

            # zero the parameter gradients
            optm.zero_grad()

            # forward + backward + optimize
            outputs = model(inputs)
            loss = crit(outputs, labels)
            loss_value = loss.item()
            loss.backward()
            optm.step()

            # Update loss totals
            running_loss += loss_value
            epoch_train_loss += loss_value

            # print statistics
            if i % batch_print_size == (batch_print_size - 1):
                print(f'Train batch {i+1}/{num_train_batches}: loss {running_loss/batch_print_size}', end='\r')
                running_loss = 0.0

        epoch_train_loss /= num_train_samples
        train_t1 = time.time() - train_t0


        valid_t0 = time.time()
        epoch_valid_loss = 0.0
        running_loss = 0.0

        with torch.no_grad():
            for i, samples in enumerate(dataloader_valid):
                # get the inputs
                inputs, labels = samples['image'], samples['bbox']
                inputs, labels = inputs.to(device), labels.to(device)

                # forward
                outputs = model(inputs)
                loss = crit(outputs, labels)
                loss_value = loss.item()

                # Update loss totals
                running_loss += loss_value
                epoch_valid_loss += loss_value

                # print statistics
                if i % batch_print_size == (batch_print_size - 1):
                    print(f'Valid batch {i+1}/{num_valid_batches}: loss {running_loss/batch_print_size}', end='\r')
                    running_loss = 0.0

        epoch_valid_loss /= num_valid_samples
        valid_t1 = time.time() - valid_t0

        epoch_train_loss_history.append(epoch_train_loss)
        epoch_valid_loss_history.append(epoch_valid_loss)

        print(f'Epoch {epoch:3d} > train: T={train_t1:5.1f}s | loss={epoch_train_loss}  valid: T={valid_t1:5.1f}s | loss={epoch_valid_loss}')

    print('Finished Training')
except KeyboardInterrupt:
    print('Finished by user')
```

### Evolução do treinamento

```python
a_epochs = np.arange(1, 1 + len(epoch_train_loss_history))
plt.plot(a_epochs, epoch_train_loss_history, 'bs', a_epochs, epoch_valid_loss_history, 'g^')
plt.show()
```

### Cálculo dos erros no conjunto de treinamento

```python
num_test_samples = len(dataloader_test.dataset)

pred_list = list()
label_list = list()
test_loss_list = list()

with torch.no_grad():
    test_loss = 0.0

    for i, samples in enumerate(dataloader_test):
        # get the inputs
        inputs = samples['image'].to(device)
        label_list.append(samples['bbox'])

        # forward
        pred_list.append(model(inputs))

test_pred = torch.cat(tuple(pred_list), 0)
print(f'Dimensão dos rótulos (bounding boxes) preditos: {test_pred.shape}')
test_label = torch.cat(tuple(label_list), 0)
print(f'Dimensão dos rótulos (bounding boxes) originais: {test_label.shape}')
```

```python
sq_test_diff = (test_pred - test_label.to(device)) ** 2
sq_test_loss = sq_test_diff.sum(1).cpu().numpy()
print(f'Dimensão dos erros quadráticos no conjunto de testes: {sq_test_loss.shape}')
test_loss_idx_list = sq_test_loss.argsort()
```

### Visualização dos resultados

```python
def vizTest(img_idxs):
    len_img_idxs = len(img_idxs)
    fig = plt.figure(figsize=(11, 4), dpi=100)
    for idx in range(len_img_idxs):
        img_idx = img_idxs[idx]
        entry = dataset_test[np.asscalar(img_idx)]

        img = entry['image'].numpy()
        img = img.transpose((1, 2, 0))
        label_bbox = test_label[img_idx].cpu().numpy()
        pred_bbox = test_pred[img_idx].cpu().numpy()

        ax = fig.add_subplot(1, len_img_idxs, 1 + idx)
        ax.imshow(img)

        h, w = img.shape[:2]
        label_bbox = [int(round(label_bbox[0] * w)),
                      int(round(label_bbox[1] * h)),
                      int(round(label_bbox[2] * w)),
                      int(round(label_bbox[3] * h))]
        pred_bbox = [int(round(pred_bbox[0] * w)),
                     int(round(pred_bbox[1] * h)),
                     int(round(pred_bbox[2] * w)),
                     int(round(pred_bbox[3] * h))]

        ax.add_patch(patches.Rectangle(label_bbox[:2], label_bbox[2], label_bbox[3],
                                       fill=False, edgecolor='blue', lw=2))
        ax.add_patch(patches.Rectangle(pred_bbox[:2], pred_bbox[2], pred_bbox[3],
                                       fill=False, edgecolor='red', lw=2))

        ax.set_title(f'{entry["label"][0]} - {sq_test_loss[img_idx]:3f}')
        ax.axis('off')
    plt.show()
```
